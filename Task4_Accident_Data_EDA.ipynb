# Exploratory Data Analysis (EDA) for Accident Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/Skillcraft Task 4 - Sheet1.csv'
df = pd.read_csv(file_path)

# Display the first 5 rows of the DataFrame
print("First 5 rows of the dataset:")
display(df.head())

# Display descriptive summary
print("\nDescriptive summary of the dataset:")
df.info()
```

## Total Accidents by Weather Condition

```python
# Calculate total accidents by 'Weather Condition' by summing 'Number of Accidents'
weather_accidents = df.groupby('Weather Condition')['Number of Accidents'].sum().reset_index()
weather_accidents.columns = ['Weather Condition', 'Total Accidents']

# Create a bar chart for 'Weather Condition'
plt.figure(figsize=(12, 7))
sns.barplot(x='Weather Condition', y='Total Accidents', data=weather_accidents, palette='viridis', hue='Weather Condition', legend=False)
plt.title('Total Accidents by Weather Condition', fontsize=16)
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Total Accidents', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Total Accidents by Road Condition

```python
# Calculate total accidents by 'Road Condition' by summing 'Number of Accidents'
road_accidents = df.groupby('Road Condition')['Number of Accidents'].sum().reset_index()
road_accidents.columns = ['Road Condition', 'Total Accidents']

# Create a bar chart for 'Road Condition'
plt.figure(figsize=(12, 7))
sns.barplot(x='Road Condition', y='Total Accidents', data=road_accidents, palette='magma', hue='Road Condition', legend=False)
plt.title('Total Accidents by Road Condition', fontsize=16)
plt.xlabel('Road Condition', fontsize=12)
plt.ylabel('Total Accidents', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Summary of Insights

Based on the Exploratory Data Analysis, we can observe the following insights regarding accident occurrences:

### Weather Condition Impact:
*   The most frequent accidents occur during **Clear** weather conditions, suggesting that while clear weather is generally safer, it accounts for the largest volume of accidents likely due to higher traffic density.
*   **Rainy** conditions also contribute significantly to accidents, indicating reduced visibility and slippery roads as major factors.
*   Accidents are less frequent in **Snow** or **Fog**, which might be due to reduced traffic volume during such severe weather, or drivers exercising greater caution.

### Road Condition Impact:
*   **Dry** road conditions account for the vast majority of accidents, similar to clear weather, likely due to higher traffic volumes.
*   **Wet** road conditions are the second most common cause of accidents, highlighting the danger of reduced traction on wet surfaces.
*   Accidents are considerably less frequent on **Snowy** or **Icy** roads, again possibly reflecting lower traffic or increased driver vigilance in such hazardous conditions.
