```python
import pandas as pd
```

### Load the dataset

```python
df = pd.read_csv('/content/Skillcraft Task 2  - train.csv')
```

### Fill missing values in 'Age' with its median and drop duplicates

```python
# Fill missing 'Age' values with the median
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

# Drop duplicate rows
df.drop_duplicates(inplace=True)
```

### Display the first 5 rows of the cleaned DataFrame

```python
display(df.head())
```

### Save the cleaned data to a new CSV file

```python
df.to_csv('Cleaned_Titanic_Data.csv', index=False)
print("Cleaned data saved as 'Cleaned_Titanic_Data.csv'")
```

### Add a dummy string date column 'Ticket_Date' and convert to datetime

```python
# Add 'Ticket_Date' column with dummy string date
df['Ticket_Date'] = '2026-05-25'

# Convert 'Ticket_Date' column to datetime format
df['Ticket_Date'] = pd.to_datetime(df['Ticket_Date'])

print("Added and converted 'Ticket_Date' column:")
display(df[['Ticket_Date']].head())
```

### Save the updated DataFrame to 'Cleaned_Titanic_Data.csv'

```python
df.to_csv('Cleaned_Titanic_Data.csv', index=False)
print("Updated data saved as 'Cleaned_Titanic_Data.csv'")
```
